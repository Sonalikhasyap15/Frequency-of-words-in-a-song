```python
import pandas as pd
import numpy as np
import matplotlib as pt
```

```python
#Importing the dataset
import os
os.chdir(r"D:\data")
```

```python
f = open("song.txt" , "r")
#Ceating a dictionary
di = dict()
#Reading each line of the datas
for line in f:
    line = line.rstrip()
    words = line.lower().replace("," , "").split()
    for word in words:
        if word in di:
            di[word] = di[word] + 1
        else:
            di[word] = 1
 
```

```python
print(di)
```

```python
#Finding the frequency of a specific word asked from user
x = input("enter the word: ")
if x in di:
    print(x,di[x])
else:
    print("sorry the word didn't exist in the song")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
```

```python
#Plotting the graph
names = list(di.keys())
values = list(di.values())
matplotlib.rcParams['font.family'] = "serif"
plt.bar(range(len(di)),values,tick_label=names)
index = np.arange(len(di))
plt.xlabel('Words',fontsize = 30)
plt.ylabel('Frequency',fontsize = 30)
plt.xticks(rotation=90)
plt.title('Word Vs Count In My Song',fontsize = 30)
plt.savefig('song_data.png', quality=100)
plt.rcParams["figure.figsize"] = (30,20)
```
